<h1>Cleaning the data</h1>

```python
pingu
```

```python
pingu['bill_length_mm'].mean()
pingu['bill_depth_mm'].mean()
pingu['flipper_length_mm'].mean()
pingu['body_mass_g'].mean()
pingu['sex'].mode()
```

```python
pingu['bill_length_mm'] = pingu['bill_length_mm'].fillna(pingu['bill_length_mm'].mean())
pingu['bill_depth_mm'] = pingu['bill_depth_mm'].fillna(pingu['bill_depth_mm'].mean())
pingu['flipper_length_mm'] = pingu['flipper_length_mm'].fillna(pingu['flipper_length_mm'].mean())
pingu['body_mass_g'] = pingu['body_mass_g'].fillna(pingu['body_mass_g'].mean())
pingu['sex'] = pingu['sex'].fillna(pingu['sex'].mode()[0])
```

```python
pingu.isnull().sum()
```

```python
pingu['species'] = pingu['species'].astype('category')
pingu['island'] = pingu['island'].astype('category')
pingu['sex'] = pingu['sex'].astype('category')
```

```python
pingu['body_mass_g'] = pingu['body_mass_g'].astype('float64')
```

```python
pingu['bill_length_mm'].mean()
pingu['bill_depth_mm'].mean()
pingu['flipper_length_mm'].mean()
pingu['sex'].mode()
```
